```python
import matplotlib.pyplot as plt
import numpy as np

# Dataset sizes and operations that we computed time one
sizes = ['2MB','1GB', '2GB', '3GB', ]  
operations = ['Load', 'Show', 'Preprocess', 'Analysis']  
times = [
        [0.417, 0.143, 1.336, 5.212],  
    [8.482, 0.219, 30.472, 8.625],  
    [17.814, 0.168, 76.142, 12.709],  
    [23.42, 0.170, 113.327, 24.804] 
]

time_values = np.array(times)  
nr = np.arange(len(sizes))  

# Colors and Bar Width
bar = 0.2
bar_colors = ['b', 'g', 'r', 'c']

plt.figure(figsize=(10, 6))

# Different bar for each operation
for i in range(len(operations)):
    plt.bar(nr + i * bar, time_values[:, i], bar, label=operations[i], color=bar_colors[i])

plt.xlabel("Cores")
plt.ylabel("Time (s)")
plt.title("Comparison of Operation Times for Different Datasets Sizes with 4 Cores")
plt.xticks(nr + bar * 1.5, sizes)  
plt.legend()
plt.grid(axis='y', linestyle="--", alpha=0.7)

plt.show()
```
